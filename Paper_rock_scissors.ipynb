# Rock Scissors Paper

This is just a little excercice to create a python script to play paper, rock, scissors game.

You need to choose between:

image.png

```python
import random
```

```python
plays = {0: 'Rock', 2: 'Scissors', 5: 'Paper'}
while True:
  machine = random.choice([0, 2, 5])
  player = 10

  while player not in [0, 2, 5]:
    #player = int(input(f'Choose your play [0, 2, 5]: '))
    player = int(input(f'Choose your play {plays}: '))

  #print(f'player: {player}  machine: {machine}')
  print(f'Player choosen: {plays[player]},  machine choosen: {plays[machine]}')

  if player == machine:
    print('Draw!')
  elif machine == 5 and player == 0:
    print('Machine wins')
  elif player == 5 and machine == 0:
    print('Player wins')
  else:
    if player < machine:
      print('Player wins!')
    else:
      print('Machine wins!')
```
